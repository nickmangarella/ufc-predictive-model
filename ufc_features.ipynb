```python
# Dependencies and Setup
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(os.path.join("Resources", "ufc-master.csv"))
df = df.fillna(df.mean())

df.head()
```

```python
plt.figure(figsize=(15, 8))

plt.subplot(2, 3, 1)
plt.hist(df['age_dif'], bins=10)
plt.title("Age Difference")

plt.subplot(2, 3, 2)
plt.hist(df['height_dif'], bins=50)
plt.title("Height Difference")

plt.subplot(2, 3, 3)
plt.hist(df['reach_dif'], bins=40)
plt.title("Reach Difference")

plt.subplot(2, 3, 4)
plt.hist(df['sig_str_dif'], bins=20)
plt.title("Significant Strike Difference")

plt.subplot(2, 3, 5)
plt.hist(df['avg_sub_att_dif'], bins=30)
plt.title("Submission Difference")

plt.subplot(2, 3, 6)
plt.hist(df['avg_td_dif'], bins=20)
plt.title("Takedown Difference")

plt.savefig("features.png", dpi=300)
```
